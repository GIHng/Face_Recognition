```python
import os
from PIL import Image

def is_image_corrupted(file_path):
    try:
        with Image.open(file_path) as img:
            img.verify()  # 이미지가 손상되었는지 확인
        return False  # 이미지가 정상적으로 로드됨
    except:
        return True  # 이미지 로드 중 오류 발생 (손상)

def delete_corrupted_images_in_directory(directory_path):
    """주어진 디렉터리 내의 모든 손상된 이미지 파일 삭제"""
    for root, dirs, files in os.walk(directory_path):
        for filename in files:
            if filename.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):  # 대상 이미지 파일 형식 지정
                file_path = os.path.join(root, filename)
                if is_image_corrupted(file_path):
                    os.remove(file_path)  # 손상된 이미지 파일 삭제
                    print(f"Deleted corrupted image: {file_path}")

# 'data' 폴더 내에서 손상된 이미지 파일 삭제    
data_directory_path = '../data'
delete_corrupted_images_in_directory(data_directory_path)
```
